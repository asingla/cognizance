```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
x = [5,8,10]
y = [12,16,6]

plt.plot(x,y)

plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

### Using Styles, they serve as CSS stylesheets for our graphs

```python
style.use('ggplot')

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

# can plot specifically, after just showing the defaults:
#changing linewidths as well
plt.plot(x,y,linewidth=5)
plt.plot(x2,y2,linewidth=5)

plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

#### Giving labels to each line and displaying legend

```python
plt.plot(x,y,label = 'value 1',linewidth=5)
plt.plot(x2,y2,label = 'value 2', linewidth=5)

plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()

plt.show()
```

#### Showing bar graphs with the same data and changing color

```python
plt.bar(x, y, align='center')

#giving a specific colour
plt.bar(x2, y2, color='g', align='center')


plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
plt.scatter(x, y)

plt.scatter(x2, y2, color='g')


plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```
