```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'titanic_train.csv' #stored in local folder, should upload on github

titanic = pd.read_csv(url)

titanic.head()
```

```python
titanic.set_index('PassengerId').head()
```

```python
def isAdult(x):
    if x>=18:
        return True
    else:
        return False
```

```python
titanic['adult'] = titanic['Age'].apply(isAdult)
titanic.head()

plt.scatter(titanic['Age'], titanic['Survived'])
plt.xlim(50,80)
plt.show()
```

```python
#titanic = titanic[titanic.Pclass == 1]
titanic.head()
```

```python
# creates the plot using
lm = sns.lmplot(x = 'Pclass', y = 'Survived', data = titanic, hue = 'Sex', fit_reg=False)

# set title
lm.set(title = 'Fare x Age')

# get the axes object and tweak it
axes = lm.axes
```

```python
# sum the instances of males and females
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

# put them into a list called proportions
proportions = [males, females]

# Create a pie chart
plt.pie(
    # using proportions
    proportions,
    
    # with the labels being officer names
    labels = ['Males', 'Females'],
    
    
    # with colors
    colors = ['blue','red'],
    
    # with one slide exploded out
    explode = (0.15 , 0),
    
    # with the start angle at 90%
    startangle = 90,
    )

# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("Sex Proportion")

# View the plot
plt.tight_layout()
plt.show()
```
